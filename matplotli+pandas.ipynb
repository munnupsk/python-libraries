# line graph

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
gas=pd.read_csv(r'C:\Users\prasanna\Desktop\gas_prices.csv')
gas
```

```python
#here we gonna plot graph for year and  usa 
```

```python
plt.title("Gas Prices")
#plt.plot(gas.Year,gas.USA,'b.-',label='United States of america')
#plt.plot(gas.Year,gas.Canada,'g.-',label='Canada')
#plt.plot(gas['Year'],gas['South Korea'],'r.-',label='South Korea')
for country in gas:
    if country!='Year':
        plt.plot(gas.Year,gas[country],marker='.')
plt.xticks(gas.Year[::3])
plt.ylabel("US dollars")
plt.xlabel("Year")
plt.legend()
plt.figure(figsize=(140,20))
plt.show()
```

# load fifa Data

```python
fifa=pd.read_csv(r'C:\Users\prasanna\Desktop\fifa_data.csv')
fifa.head(5)
```

# histogram

```python
plt.title('Distribution table in FIFA 2018')
plt.hist(fifa.Overall,bins=[40,50,60,70,80,90,100],color='#264577')
plt.xlabel("Skill Level")
plt.ylabel("Number of Player")
plt.show()
```

# pie chart

```python
plt.title("Foot preference of football players")
left=fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]
right=fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]
labels=['left','right']
colors=['#899ab5','#b1bcce']
plt.pie([left,right],labels=labels,colors=colors,autopct='%.2f %%')
plt.show()
```

```python
plt.title('professional soccer team')
barcelona=fifa.loc[fifa.Club=='FC Barcelona'] ['Overall']
madrid=fifa.loc[fifa.Club=='Real Madrid']['Overall']
plt.boxplot([barcelona,madrid],labels=['Barcelona','Red Madrid'])
plt.show()
```

```python
plt.title('professional soccer team')
barcelona=fifa.loc[fifa.Club=='FC Barcelona'] ['Overall']
madrid=fifa.loc[fifa.Club=='Real Madrid']['Overall']
boxes=plt.boxplot([barcelona,madrid],labels=['Barcelona','Red Madrid'],patch_artist=True)
for box in boxes['boxes']:
    box.set(color='#4286f4',linewidth=2)
    box.set(facecolor='#e0e0e0')
plt.show()
```
